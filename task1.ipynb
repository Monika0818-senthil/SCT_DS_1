```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a sample dataset
data = {
    'Age Group': ['0 to 20 Years', '21 to 64 Years', '65+ Years'],
    'Population (Millions)': [512, 807, 98],
    'Percentage': [36.1, 57.0, 6.9]
}

df = pd.DataFrame(data)

# Step 2: Plotting
age_groups = df['Age Group']
population_mn = df['Population (Millions)']
percentages = df['Percentage']
colors = ['gold', 'dodgerblue', 'deeppink']

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(age_groups, population_mn, color=colors)

# Add labels to each bar
for bar, pct, pop in zip(bars, percentages, population_mn):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10,
             f'{pop} Mn\n({pct}%)',
             ha='center', va='bottom', fontsize=12, fontweight='bold')

# Customize plot
plt.title("India's Population Distribution by Age in 2022", fontsize=14, fontweight='bold')
plt.ylabel("Population (in millions)")
plt.ylim(0, max(population_mn) + 100)

# Add annotation for total population
plt.text(1, max(population_mn) + 50,
         'Total Population: 1.42 Billion\nMedian Age: 28',
         ha='center', fontsize=12,
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round'))

plt.tight_layout()
plt.show()
```
